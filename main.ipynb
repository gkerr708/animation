```python
import numpy as np, matplotlib.pyplot as plt, random
from matplotlib import animation
```

```python
x, y = [], []
for i in range(100):
    x.append(random.randint(0,100))
    y.append(random.randint(0,100))
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    plt.scatter(x,y, color='blue')
    plt.pause(0.0005)
plt.show()
```
